```python
# Libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Synthetic Thanksgiving Data
np.random.seed(2310)    # set seed to 2310
n = 50    # set n to 50

# creating the variables first (Age, Dinner, Dessert, Duration)
Age = (np.random.normal(37, 18, n).round(0).clip(5, 90))
Num_Dinner_Servings = np.round(4 - 0.0009 * (Age - 40)**2 + np.random.normal(0, 0.3, n)).clip(1, 5)
Num_Dessert_Servings = np.round(np.clip(2.5 - (Age / 50) + np.random.normal(0, 0.5, n), 0, 3))
Dinner_Duration_Hours =  np.round(1.5 + (Age / 40) + np.random.normal(0, 0.5, n), 1).clip(1, 5)

# creating boolean for whether or not they watched tv (older and younger people more likely to watch)
prob_parade = np.clip(    0.3 + 0.4 * np.exp(-((Age - 15)/20)**2) + 0.4 * np.exp(-((Age - 70)/20)**2),0, 1)
Watched_Parade = np.random.binomial(1, prob_parade)

# creating the data frame using the variables we just made
Thanksgiving = pd.DataFrame({
    'Ages' : Age,
    'Dinner_Servings' : Num_Dinner_Servings,
    'Dessert_Servings' : Num_Dessert_Servings,
    'Duration' : Dinner_Duration_Hours,
    'Macy_Parade' : Watched_Parade
})

Thanksgiving.head()
```

```python
avgduration = (Thanksgiving.groupby(['Ages','Macy_Parade'])['Duration'].mean().unstack())

avgduration.plot(
    kind='bar',
    stacked=True,
    figsize=(12, 6),
    edgecolor='brown',
    color=['red', 'orange']
)

plt.xlabel('Age')
plt.ylabel('Average Duration')
plt.title('Average Duration by Age and Macy Parade')
plt.legend(['No', 'Yes'], title='Watched Macy Parade', labelcolor='brown')
plt.show()
```

### Graph Description:
> Here we can see avg duration by age group along with the amount per group who watched the Macy's Parade. We can see that many of the age groups did end up watching the Macys Parade, we can see that on each end ranges 9-19 and 55-72 did infact all watch the parade which tells us that it is not a matter of young or old.
